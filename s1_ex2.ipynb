```python
# Program: Address Formatter
# Purpose: Collect address details from the user and display them

def get_user_address():
    """Collect address details from the user and return them."""

    # Input details
    full_name = input("Enter your full name: ")
    house_number = input("Enter your house number: ")
    city_name = input("Enter your city: ")
    state_name = input("Enter your state: ")
    postal_code_value = input("Enter your postal code: ")

    # IMPORTANT: return the collected values
    return full_name, house_number, city_name, state_name, postal_code_value


def print_address(name, house, city, state, postal_code):
    """Print the formatted address."""
    print("\nFormatted Address:")
    print(f"{name}\n{house}\n{city}, {state} - {postal_code}")


# Main execution
user_name, user_house, user_city, user_state, user_postal = get_user_address()
print_address(user_name, user_house, user_city, user_state, user_postal)
```
